🧩 Exercício 9: Herança — Funcionários

Crie:

Classe base:

 - Funcionario (nome, salário)

Classe filha:

 - Gerente (bônus adicional)

Método:

1 - calcular salário total

```python
class Funcionario:
    def __init__(self, nome, salario):
        self.nome    = nome
        self.salario = salario

    def calcular_salario(self):
        return self.salario

class Gerente(Funcionario):
    def __init__(self, nome, salario, bonus):
        super().__init__(nome, salario) #herda da classe pai
        self.bonus = bonus

    def calcular_salario(self):
        return self.salario + self.bonus
```

```python
func = Funcionario("João", 3000)
ger = Gerente("Maria", 5000, 2000)

print(func.calcular_salario())
print(ger.calcular_salario())
```
