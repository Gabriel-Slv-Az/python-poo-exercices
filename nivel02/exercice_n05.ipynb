🧩 Exercício 5: Aluno

Crie uma classe Aluno com:

* nome
* lista de notas

Métodos:

1 - adicionar nota

2 - calcular média

3 - verificar se foi aprovado (média >= 7)

```python
class Aluno:
    def __init__(self, nome, nota_inicial: float):
        self.nome = nome
        self._notas = [nota_inicial]

    def adicionar_notas(self, nota_prova: float):
        self._notas.append(nota_prova)

    def calcula_media(self):
        return sum(self._notas) / len(self._notas)

    @property
    def notas(self):
        return list(self._notas)
```

```python
aluno1 = Aluno("Felipe", 10)
```

```python
aluno1.adicionar_notas(9)
```

```python
aluno1.calcula_media()
```
