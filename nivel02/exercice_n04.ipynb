🧩 Exercício 4: Conta Bancária

Crie uma classe ContaBancaria com:

* saldo

* Métodos:

1 - depositar(valor)

2 - sacar(valor) (não pode sacar mais que o saldo)

3 - ver_saldo()

```python
class ContaBancaria:
    def __init__(self, saldo: float):
        self.saldo = saldo

    def depositar(self, valor: float):
        if valor <= 0:
            raise ValueError("Valor de depósito deve ser positivo")
        self.saldo += valor

    def sacar(self, saque: float):
        if self.saldo < saque:
            return False
        self.saldo -= saque
        return True

    def saldo_atual(self):
        return self.saldo
```

```python
cta1 = ContaBancaria(1000)
```

```python
cta1.depositar(500)
```

```python
cta1.saldo_atual()
```

```python
cta1.sacar(5000)
```

```python
cta1.sacar(500)
```

```python
cta1.saldo_atual()
```
