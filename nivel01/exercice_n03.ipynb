🧩 Exercício 3: Contador

Crie uma classe Contador que:

* começa em 0
* tenha um método incrementar()
* tenha um método valor() que retorna o valor atual

```python
class Contador:
    def __init__(self, contagem: int):
        self.contagem = contagem

    def incrementar(self):
        self.contagem += 1
        return self.contagem

    def valor(self):
        return self.contagem
```

```python
contagem1 = Contador(10)
```

```python
contagem1.incrementar()
```

```python
contagem1.valor()
```
