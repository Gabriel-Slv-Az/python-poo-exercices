🧩 Exercício 2: Classe Pessoa

Crie uma classe Pessoa com:

* nome
* idade

Crie um método que:

diga se a pessoa é maior de idade (retorne True/False)

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome  = nome
        self.idade = idade

    def valid_maior_idade(self):
        return self.idade >= 18

pessoa1 = Pessoa("Gabriel", 22)
pessoa2 = Pessoa('Julia', 17)
```

```python
pessoa1.valid_maior_idade()
```

```python
pessoa2.valid_maior_idade()
```
