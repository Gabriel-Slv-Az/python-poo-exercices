🧩 Exercício 1: Classe Produto

Crie uma classe Produto com:

* nome
* preço

Crie um método que:
retorna o preço com 10% de desconto

```python
class Produto:
    def __init__(self, nome, preco):
        self.nome  = nome
        self.preco = preco

    def desconto(self, desconto):
        return self.preco - (self.preco * desconto)
```

```python
p1 = Produto("Notebook", 2000)
```

```python
p1_descontado = p1.desconto(0.1)
```
