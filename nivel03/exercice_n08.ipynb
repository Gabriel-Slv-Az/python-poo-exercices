🧩 Exercício 8: Controle de estoque

Classe Estoque:

* produto
* quantidade

Métodos:

1 - adicionar estoque

2 - remover estoque (não pode ficar negativo)

```python
class Estoque:
    def __init__(self, produto: str, quantidade: int):
        self.produto = produto
        self.quantidade = quantidade

    def __repr__(self):
        return f"Estoque(produto={self.produto!r}, quantidade={self.quantidade})"

class ControleEstoque:
    def __init__(self):
        self._itens: dict[str, Estoque] = {}

    def adicionar_estoque(self, produto: str, quantidade: int) -> None:
        if quantidade <= 0:
            raise ValueError("Quantidade deve ser positiva")

        if produto in self._itens:
            self._itens[produto].quantidade += quantidade
        else:
            self._itens[produto] = Estoque(produto, quantidade)

    def remover_estoque(self, produto: str, quantidade: int) -> bool:
        if quantidade <= 0:
            raise ValueError("Quantidade deve ser positiva")

        if produto not in self._itens:
            raise ValueError("Produto não encontrado")

        item = self._itens[produto]

        if item.quantidade < quantidade:
            raise ValueError("Estoque insuficiente")

        item.quantidade -= quantidade

    @property
    def estoque_atual(self) -> dict[str, Estoque]:
        return self._itens
```

```python
controle_estoque1 = ControleEstoque()
```

```python
controle_estoque1.adicionar_estoque("Calca", 500)
```

```python
controle_estoque1.estoque_atual
```

```python
controle_estoque1.remover_estoque("Calca", 250)
```
