🧩 Exercício 7: Conta segura

Melhore a ContaBancaria:

1 - saldo deve ser “privado” (_saldo)

2 - só pode ser alterado via métodos

```python
class ContaBancaria:
    def __init__(self, saldo: float):
        self._saldo = saldo

    def depositar(self, valor: float) -> None:
        if valor <= 0:
            raise ValueError("Valor de depósito deve ser positivo")
        self._saldo += valor

    def sacar(self, saque: float) -> bool:
        if saque <= 0:
            raise ValueError("Valor de saque deve ser positivo")
        if self._saldo < saque:
            return False
        self._saldo -= saque
        return True

    @property
    def saldo_atual(self: float):
        return self._saldo
```

```python
conta1 = ContaBancaria(500)
```

```python
conta1.depositar(500)
```

```python
conta1.saldo_atual
```

```python
conta1.sacar(500.00)
```

```python
conta1.saldo_atual
```
